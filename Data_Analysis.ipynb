```python
import pandas as pd

names = ['gender', 'hypertension', 'heart_disease', 'ever_married', 'work_type', 'Residence_type', 'smoking_status']

for name in names:
    print(name)
    print("eil skaicius:", df[name].count())

    print("trukstamos reiksmes %:", df[name].isnull().mean() * 100)

    print("?Kardinalumas:", df[name].nunique())

    mode_value = df[name].mode()[0] if not df[name].mode().empty else None
    print("Moda:", mode_value)

    mode_frequency = df[name].value_counts().iloc[0] if not df[name].value_counts().empty else None
    print("Modos daznumas:", mode_frequency)

    mode_percentage = (mode_frequency / df[name].count()) * 100 if df[name].count() > 0 else None
    print("Moda %:", mode_percentage)

    second_mode_value = df[name].value_counts().index[1] if len(df[name].value_counts()) > 1 else None
    print("antroji moda:", second_mode_value)

    second_mode_frequency = df[name].value_counts().iloc[1] if len(df[name].value_counts()) > 1 else None
    print("antrosios modos daznumas:", second_mode_frequency)

    second_mode_percentage = (second_mode_frequency / df[name].count()) * 100 if df[name].count() > 0 and second_mode_frequency is not None else None
    print("antroji moda %:", second_mode_percentage)

    print("\n" + "-"*50 + "\n")
```

```python

continuous_vars = ['avg_glucose_level', 'bmi']

def count_outliers(df, column):
    Q1 = df[column].quantile(0.25)  
    Q3 = df[column].quantile(0.75)  
    IQR = Q3 - Q1  
    lower_bound = Q1 - 1.5 * IQR  
    upper_bound = Q3 + 1.5 * IQR  
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)] 
    return len(outliers)

outliers_counts = {col: count_outliers(df, col) for col in continuous_vars}

print("Outlier Counts:")
for key, value in outliers_counts.items():
    print(f"{key}: {value} outliers")
```

```python

continuous_vars = ['age', 'avg_glucose_level', 'bmi']


plt.figure(figsize=(12, 8))

plot_index = 1
for i in range(len(continuous_vars)):
    for j in range(i + 1, len(continuous_vars)):
        plt.subplot(2, 2, plot_index)  
        
        sns.regplot(x=df[continuous_vars[i]], y=df[continuous_vars[j]], scatter=True, line_kws={"color": "red"})
        
        plt.xlabel(continuous_vars[i])
        plt.ylabel(continuous_vars[j])
        plt.title(f"{continuous_vars[i]} ir {continuous_vars[j]}")
        
        correlation = np.corrcoef(df[continuous_vars[i]], df[continuous_vars[j]])[0, 1]
        
        plt.text(0.05, 0.9, f"Corr: {correlation:.2f}", transform=plt.gca().transAxes, 
                 fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
        
        plot_index += 1

plt.tight_layout()
plt.show()
```

```python

categorical_vars = ['hypertension', 'gender']

fig, axes = plt.subplots(1, 3, figsize=(15, 5))

def add_count_labels(ax, data):
    for p in ax.patches:
        ax.annotate(f'{p.get_height():.2f}', 
                    (p.get_x() + p.get_width() / 2, p.get_height()), 
                    ha='center', va='bottom', fontsize=10, color='black')

ax1 = sns.barplot(x=df['hypertension'].value_counts(normalize=True).index, 
                  y=df['hypertension'].value_counts(normalize=True).values, 
                  ax=axes[0])
axes[0].set_title("Bendras pasiskirstymas")
axes[0].set_xlabel("Hypertension")
axes[0].set_ylabel("Danis")
add_count_labels(ax1, df)

ax2 = sns.barplot(x=df[df['gender'] == 'Male']['hypertension'].value_counts(normalize=True).index, 
                  y=df[df['gender'] == 'Male']['hypertension'].value_counts(normalize=True).values, 
                  ax=axes[1])
axes[1].set_title("Pasiskirstymas tarp vyr")
axes[1].set_xlabel("Hypertension")
axes[1].set_ylabel("Danis")
add_count_labels(ax2, df[df['gender'] == 'Male'])

ax3 = sns.barplot(x=df[df['gender'] == 'Female']['hypertension'].value_counts(normalize=True).index, 
                  y=df[df['gender'] == 'Female']['hypertension'].value_counts(normalize=True).values, 
                  ax=axes[2])
axes[2].set_title("Pasiskirstymas tarp moter")
axes[2].set_xlabel("Hypertension")
axes[2].set_ylabel("Danis")
add_count_labels(ax3, df[df['gender'] == 'Female'])

# Adjust layout
plt.tight_layout()
plt.show()
```

```python


continuous_vars = ['age', 'avg_glucose_level', 'bmi']


def min_max_normalization(x, a=0, b=1):
    return ((x - x.min()) / (x.max() - x.min())) * (b - a) + a

for col in continuous_vars:
    df[col] = min_max_normalization(df[col], a=-1, b=1)  

df.head(20)
```

```python

from sklearn.preprocessing import LabelEncoder

categorical_features = ["gender", "ever_married", "work_type", "Residence_type", "smoking_status"]


label_encoders = {} 
for col in categorical_features:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])  
    label_encoders[col] = le  

df.head(20)
```
